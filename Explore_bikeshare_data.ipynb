### Question 1


**Trip duration.**

```r
# What is the total travel time for users in different cities?

# Function to calculate the total:
totalTravelTime = function(df){
    total = sum(df$Trip.Duration, na.rm=TRUE)
    return(total)
}

# New York City
NewYork = totalTravelTime(ny)

# Washington
Washington = totalTravelTime(wash)

# Chicago
Chicago = totalTravelTime(chi)

# Results
cat("New York city users total travel time is: ", NewYork, "\n" )
cat("Washington users total travel time is: ", Washington, "\n" )
cat("Chicago users total travel time is: ", Chicago, "\n" )

# Values are divide by 1000000 for readabilty

# Plotting the graph


library(ggplot2)
data <- data.frame(
  City_Name=c("New York City","Washington","Chicago") ,  
  Total_Travel_Time=c(49.490073,109.883545,8.087801)
  )
ggplot(data, aes(x=City_Name, y=Total_Travel_Time)) + 
geom_bar(stat = "identity", width=0.5)
```

```r
# What is the average travel time for users in different cities?

# Function to calculate average
averageTravelTime = function(df) {
    average = round(mean(df$Trip.Duration, na.rm=TRUE), digits = 2)
    return(average)
}

# New York City Average
NewYorkAverage = averageTravelTime(ny)

# Washington Average
WashingtonAverage = averageTravelTime(wash)

# Chicago Average
ChicagoAverage = averageTravelTime(chi)

# Results
cat("New York city average travel time for users is: ", NewYorkAverage, "\n" )
cat("Washington average travel time for users is: ", WashingtonAverage, "\n" )
cat("Chicago average travel time for users is: ", ChicagoAverage, "\n" )

# Values divided by 100 for readability

# Plotting the graph

library(ggplot2)
data <- data.frame(
  City_Name=c("New York City","Washington","Chicago") ,  
  Average_Travel_Time=c(9.03,12.34,9.37)
  )
 ggplot(data, aes(x=City_Name, y=Average_Travel_Time)) + 
 geom_bar(stat = "identity", width=0.3) +
  coord_flip()
```

**Bikers from Washington seem to prefer the bikeshare method of travelling from point A to point B followed by New York City then Chicago. Though in Chicago this does not seem to be the most used method of sharing, bikers in that city seem to spend more average time travelling compared to bikers in New York.**

```r
# What is the most common trip from start to end (i.e., most frequent combination of start station and end station)?

combinations = function(df){
    combination = paste(df$Start.Station,df$End.Station, sep=" and ")
    return(combination)
}

trips =  function(df){
    trip = tail(names(sort(table(combinations(df)))), 1)
    return (trip)
}

newYorkTrip = trips(ny)
washingtonTrip = trips(wash)
chicagoTrip = trips(chi)

# Results
cat("Most frequent trip combinations \n")
cat("New York City: ", newYorkTrip, "\n" )
cat("Washington: ", washingtonTrip, "\n" )
cat("Chicago: ", chicagoTrip, "\n" )

```

**Bikers seem to prefer same spot as either their starting point or ending point. That is why it is not surprising to see that the most common trips from start to end were actually round trips.**
